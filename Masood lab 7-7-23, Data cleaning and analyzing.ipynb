# Data Cleaning: 
For better analysis and visulaization we use regex function to clean the icome column, the income column have intergers and numeric, first we delete the $ sign and comma (,) from the colmun

```python
us_census['Income'] = us_census['Income'].replace('[\$,]', '', regex=True)
us_census['Income'] = pd.to_numeric(us_census['Income'])

#A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.
#RegEx can be used to check if a string contains the specified search pattern.
#import re (to call the regular expression module)
```

```python
us_census.head()
```

# second step of cleaning
here in GenderPop column, male and female attributes are in the same cell, it is therefore  necessary to split the Male and Female. because we have to analyse only female datatype

```python
us_census[['Male', 'Female']] = us_census['GenderPop'].str.split('_', expand=True)
```

```python
us_census.head()
```

```python
us_census['Male'] = us_census['Male'].str[:-1]
us_census['Female'] = us_census['Female'].str[:-1]

#remove the string "M" and "F" from the male and females columns
```

```python
us_census.head()
```

```python
us_census['Male'] = pd.to_numeric(us_census['Male'])
us_census['Female'] = pd.to_numeric(us_census['Female'])

# convert into numeric using pandas library
```

```python
us_census.head()
```

```python
#fill null values if any

us_census['Female'] = us_census['Female'].fillna(us_census['TotalPop'] - us_census['Male'])
```
